# **<font color="#004D7F" size=6> 1. ¿Qué es Lazypredict?</font>**

- **Lazypredict** es una biblioteca de Python que permite entrenar y evaluar rápidamente múltiples modelos de Machine Learning con el mínimo esfuerzo.

- Su objetivo principal es facilitar la comparación de distintos algoritmos sin necesidad de realizar una extensa configuración manual, lo que resulta especialmente útil en la fase exploratoria de un proyecto.

- Funciona tanto en tareas de **regresión** como de **clasificación**.

- Software libre: licencia MIT

# **<font color="#004D7F" size=6> 2. Características principales</font>**

1. **Entrenamiento rápido:** Ejecuta múltiples modelos de aprendizaje automático con solo unas pocas líneas de código.

2. **Comparación automática:** Genera un resumen con métricas de rendimiento (Accuracy, RMSE, R², etc.) para facilitar la selección del mejor modelo.

3. **Compatibilidad:** Funciona con scikit-learn y es fácil de integrar en flujos de trabajo de Machine Learning.

4. **Facilidad de uso:** No requiere ajustar hiperparámetros ni realizar preprocesamiento avanzado.

# **<font color="#ad0ca8" size=5> Modelos de Regresión</font>**

```python
# Importar las librerias necesarias
from lazypredict.Supervised import LazyRegressor
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# Cargar el dataset de datos inmobiliarios de Boston
data_url = "https://github.com/selva86/datasets/raw/master/BostonHousing.csv"
boston_df = pd.read_csv(data_url)

# Seleccionar las características y variable objetivo
X = boston_df[['crim', 'indus', 'nox', 'rm', 'age', 'dis', 'tax', 'ptratio', 'b', 'lstat']]
y = boston_df['medv']

# Dividir los datos en conjunto de entrenamiento (80%) y prueba (20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2025)

# Instanciar LazyRegressor para probar múltiples modelos de regresión
reg = LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)

# Entrenar y evaluar modelos automáticamente en los datos de entrenamiento y prueba
models, predictions = reg.fit(X_train, X_test, y_train, y_test)

# Convertir los resultados a un DataFrame de pandas para visualizar mejor la tabla
models_df = pd.DataFrame(models)
```

```python
# Mostrar la tabla con los resultados
models_df
```

```python
# Plot del ranking de modelos
import matplotlib.pyplot as plt
import seaborn as sns
predictions["RMSE"] = [0 if i < 0 else i for i in predictions.iloc[:,0] ]
plt.figure(figsize=(5, 10))
sns.set_theme(style="whitegrid")
ax = sns.barplot(y=predictions.index, x="RMSE", data=predictions)
ax.set(xlim=(0, 1))
```

# **<font color="#ad0ca8" size=5> Modelos de Clasificación</font>**

```python
# Importar las librerias necesarias
import pandas as pd
import numpy as np
from lazypredict.Supervised import LazyClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

# Cargar el dataset de datos inmobiliarios de Boston
data_url = "https://raw.githubusercontent.com/barzansaeedpour/cancer-detection/main/data.csv"
cancer_df = pd.read_csv(data_url)

# Seleccionar las características y variable objetivo
X = cancer_df[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean']]
y = cancer_df['diagnosis']

# Dividir los datos en conjunto de entrenamiento (80%) y prueba (20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2025)

# Instanciar LazyRegressor para probar múltiples modelos de regresión
clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)

# Entrenar y evaluar modelos automáticamente en los datos de entrenamiento y prueba
models, predictions = clf.fit(X_train, X_test, y_train, y_test)

# Convertir los resultados a un DataFrame de pandas para visualizar mejor la tabla
models_df = pd.DataFrame(models)
```

```python
# Mostrar la tabla con los resultados
models_df
```

```python
# Plot del ranking de modelos
import matplotlib.pyplot as plt
import seaborn as sns
predictions["F1 Score"] = [0 if i < 0 else i for i in predictions.iloc[:,0] ]
plt.figure(figsize=(5, 10))
sns.set_theme(style="whitegrid")
ax = sns.barplot(y=predictions.index, x="F1 Score", data=predictions)
ax.set(xlim=(0, 1))
```

# **<font color="#004D7F" size=6> 6. Recomendaciones</font>**




- Lazypredict **no realiza tratamiento de datos** como imputación de valores faltantes o normalización de variables de manera automática.

- Lazypredict **no optimiza los hiperparámetros** y solo usa los valores predeterminados de scikit-learn. Para mejorar el rendimiento de los modelos, se recomienda hacer ajuste de hiperparámetros manual o automático después de identificar los modelos más prometedores.
